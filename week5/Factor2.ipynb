```python
data_path = '/mnt/data/data/main/tick/rb'
files = sorted(os.listdir(data_path))
data_list = [pd.read_csv(f"{data_path}/{file}") for file in files]
df = pd.concat(data_list).set_index('datetime').sort_index()
```

```python
def bid_ask_imblance(bid_price, bid_volume, ask_price, ask_volume):
    return (bid_volume - ask_volume) / (bid_volume + ask_volume) / (ask_price - bid_price) / (ask_price + bid_price)
```

```python
sub_df = df.iloc[:10000]
```

```python
sub_df['bid_ask_imbalance'] = (sub_df.bid_volume1 - sub_df.ask_volume1)/ (sub_df.bid_volume1 + sub_df.ask_volume1) / (sub_df.ask_price1 - sub_df.bid_price1) / (sub_df.ask_price1 + sub_df.bid_price1)
```
