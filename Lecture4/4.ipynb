```python
import pandas as pd
import json, os, time
pd.set_option('display.max_rows', 2000)


def ticker2kline(path_list):
    for path in path_list:
        
        df = pd.read_csv(path, encoding='gb2312')
        symbol = df.合约代码[0]
        date = df.业务日期[-1]
        dir_path = os.path.join(symbol)
        file_path_bid = os.path.join(symbol, date + '_bid')
        file_path_ask = os.path.join(symbol, date + '_ask')
        
        df.index = pd.to_datetime(df.业务日期.astype(str) + pd.Series([" "] * len(df)) + df.最后修改时间.astype(str) + pd.Series([" "] * len(df)) + df.最后修改毫秒.astype(str), format="%Y%m%d %H:%M:%S %f")
        bid_kline = df['申买价一'].resample('1min', closed='right').ohlc()
        ask_kline = df['申卖价一'].resample('1min', closed='right').ohlc()
        
        if not os.path.exists(dir_path):
            os.mkdir(dir_path)
            
        bid_kline.to_pickle(file_path_bid)
        ask_kline.to_pickle(file_path_ask)
```
