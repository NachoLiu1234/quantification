```python
def get_trade_time(symbol):
    """ 
    Return trade datettime includes: night/day open, close datetime
    if no break in morning, moring break start/end datetime would be set as monring_trade_end_time
    (一)大连、上海、郑州交易所
 
    集合竞价申报时间：08：55—08：59
    集合竞价撮合时间：08：59—09：00
    正常开盘交易时间：09：00－11：30 （小节休息10：15－10：30）
    13：30－15：00

    提示：客户下单时间为集合竞价时间和正常交易时间。在8：59—9：00竞价结束时间和交易所小节休息时间（上午10:15-10:30）下单，交易系统将不接受指令，并视之为废单。（时间以交易所时钟报时为准）

    (二)上期所夜盘

    集合竞价申报时间：20：55—20：59
    集合竞价撮合时间：20：59—21：00
    正常开盘交易时间：21：00－02：30 （黄金、白银）
    21：00－01：00 （铜、铝、铅、锌、镍、锡）
    21：00－23：00（螺纹钢、热轧卷板、石油沥青、天然橡胶）
    """
    if symbol in {'黄金', '白银'}:
        night_trade_start_datetime, night_trade_end_datetime = (21, 0, 0), (2, 30, 0)
    if symbol in {'铜', '铝', '铅', '锌', '镍', '锡'}:
        night_trade_start_datetime, night_trade_end_datetime = (21, 0, 0), (1, 0, 0)
    if symbol in {'螺纹钢', '热轧卷板', '石油沥青', '天然橡胶'}:
        night_trade_start_datetime, night_trade_end_datetime = (21, 0, 0), (23, 0, 0)
    
    morning_trade_start_datetime = (9, 0, 0)
    morning_break_start_datetime = (10, 15, 0)
    morning_break_end_datetime = (10, 30, 0)
    morning_trade_end_datetime = (11, 30, 0)
    afternoon_trade_start_datetime = (13, 30, 0)
    afternoon_trade_end_datetime = (15, 0, 0)
    
    return (night_trade_start_datetime, night_trade_end_datetime, morning_trade_start_datetime,
    morning_break_start_datetime, morning_break_end_datetime, morning_trade_end_datetime,
    afternoon_trade_start_datetime, afternoon_trade_end_datetime)
```

```python
def tick2min(df, trade_time=[[(21, 0, 0), (23, 0, 0)], [(9, 0, 0), (10, 15, 0)], [(10, 30, 0), (11, 30, 0)], [(13, 30, 0), (15, 0, 0)]]):
    """
    Input: a raw DataFrame of market tick data
    Returns: a DataFrame with Open High Low Close price sampled by 1 minute
    resample
    """
    def add_1_min(lis):  # [23, 59, 59] -> [0, 0, 0] [12, 23, 50] -> [12, 23, 51]
        lis[2] += 1
        if lis[2] == 60:
            lis[1] += 1
            lis[2] = 0
        if lis[1] == 60:
            lis[0] += 1
            lis[1] = 0
        if lis[0] == 24:
            lis[0] = 0
        return lis
    
    def trade_date2ts(trade_date):  # 20200228
        return time.mktime(time.strptime(str(trade_date), '%Y%m%d'))
    
    def time2ts(hms):  # 09:00:02
        h, m, s = [int(el) for el in hms.split(':')]
        return h * 3600 + m * 60 + s    
    
    def is_trading_time(ticker, all_trade_time):
        ts = ticker[-2]
        t = time.localtime(ts)
        tup = (t.tm_hour, t.tm_min, t.tm_sec, int(ts % 1 * 1000))
        if tup in all_trade_time:
            return True
        else:
            return False
    
    all_trade_time = []
    for f, t in trade_time:  # (21, 0, 0), (2, 30, 0)
        f, t = list(f), list(t)
        while f != t:
            all_trade_time.append(tuple(f + [500]))
            f = add_1_min(f)
            all_trade_time.append(tuple(f + [0]))
    all_trade_time = set(all_trade_time)
    
    df['end_ts'] = df['trade_date'].apply(trade_date2ts) + df['time'].apply(time2ts) + (df['millisecond'] / 1000)
    df['from_ts'] = df['end_ts'] - 0.5
    
    data = list(zip(df['open'], df['high'], df['low'], df['from_ts']))    
    data = {el[3]: el for el in data}
    data_ts = set(data.keys())
    
    from_ts = min(all_ts)
    to_ts = max(all_ts)
    all_ts = {i / 2 for i in range(int(from_ts * 2), int(to_ts * 2 + 1))}
    
    # fill the void tick
    void_ts = list(all_ts - data_ts).sort()
    for ts in void_ts:
        data[ts] = data[ts - 0.5][:3] + [ts]
    
    # add close
    for i, ts in enumerate(sorted(list(all_ts))):
        try:
            data[ts].append(data[ts + 0.5][0])
        except KeyError:
            pass
    
    # filter the ticker in trading time
    data = filter(is_trading_time, data)
    
    return data
```
